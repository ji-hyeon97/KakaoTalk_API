```python
import requests, re, json
url = "https://kauth.kakao.com/oauth/token"
d = {
    'grant_type': 'authorization_code',
     'client_id': 'rest_api_key',
     'redirect_url': 'redirect_uri',
     'code': 'authorize_code'
}
response = requests.post(url, data=d)
tokens = response.json()
```

```python
fh = open('access_token.txt','w')
fh.write(tokens['access_token'])
fh.close()

fh = open('refresh_token.txt','w')
fh.write(tokens['refresh_token'])
fh.close()
```

```python
fh = open('access_token.txt','r')
access_token = fh.read()
fh.close()

url = 'https://kapi.kakao.com/v2/api/talk/memo/default/send'
h = {'Authorization':f'Bearer {access_token}'}
ttt = '메세지 보내주세요! http://dankook.ac.kr'
t = json.dumps({'object_type' : 'text', 'text':ttt, 'link':{'web_url':'redirect_uri'}})
d = {'template_object':t}
response = requests.post(url, headers=h, data=d)
#print(response.text)
```

```python
fh = open('refresh_token.txt','r')
refresh_token = fh.read()
fh.close()
url = "https://kauth.kakao.com/oauth/token"
d = {
    'grant_type': 'refresh_token',
     'client_id': 'rest_api_key',
     'redirect_url': 'redirect_uri',
     'refresh_token': refresh_token
}
response = requests.post(url, data=d)
tokens = response.json()
#print(tokens)
```
